```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined dataset
df = pd.read_csv("combined_churn_dataset.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns) 

# Convert TRUE/FALSE to 1/0 
if df['churn'].dtype == 'object':
    df['churn'] = df['churn'].map({'TRUE':1, 'FALSE':0, 'True':1, 'False':0})
```

```python
# Bar plot
df['churn'].value_counts().plot(kind='bar')
plt.title("Churn Count")
plt.show()
```

```python
# Line plot 
df['total_day_minutes'].head(50).plot()
plt.title("Day Minutes Trend")
plt.show()
```

```python
# Scatter plot
plt.scatter(df['total_day_minutes'], df['total_day_charge'])
plt.xlabel("Minutes")
plt.ylabel("Charge")
plt.title("Minutes vs Charge")
plt.show()
```
