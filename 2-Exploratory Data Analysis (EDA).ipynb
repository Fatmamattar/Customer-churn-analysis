```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined dataset
df = pd.read_csv("combined_churn_dataset.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns) 

# Convert TRUE/FALSE to 1/0 
if df['churn'].dtype == 'object':
    df['churn'] = df['churn'].map({'TRUE':1, 'FALSE':0, 'True':1, 'False':0})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mean, Median, Std
print(df.describe())

# Median separately
print("\nMedian:\n", df.median(numeric_only=True))

# Mode
print("\nMode:\n", df.mode().iloc[0])
```

```python
# Histogram Distribution
df.hist(figsize=(12,10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
#Box plot for outliers
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['total_day_minutes','total_eve_minutes','total_night_minutes']])
plt.title("Boxplot of Usage Minutes")
plt.show()
```

```python
#scatter plot for relationships
plt.scatter(df['total_day_minutes'], df['total_day_charge'])
plt.xlabel("Total Day Minutes")
plt.ylabel("Total Day Charge")
plt.title("Minutes vs Charge")
plt.show()
```
