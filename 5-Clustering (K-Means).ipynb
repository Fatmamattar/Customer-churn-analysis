```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined dataset
df = pd.read_csv("combined_churn_dataset.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns) 

# Convert TRUE/FALSE to 1/0 
if df['churn'].dtype == 'object':
    df['churn'] = df['churn'].map({'TRUE':1, 'FALSE':0, 'True':1, 'False':0})
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df[['total_day_minutes', 'total_eve_minutes', 'total_night_minutes']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Plot
sns.scatterplot(x=df['total_day_minutes'], 
                y=df['total_eve_minutes'], 
                hue=df['cluster'])
plt.show()
```
