```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined dataset
df = pd.read_csv("combined_churn_dataset.csv")

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns) 

# Convert TRUE/FALSE to 1/0 
if df['churn'].dtype == 'object':
    df['churn'] = df['churn'].map({'TRUE':1, 'FALSE':0, 'True':1, 'False':0})
```

```python
print(df.dtypes)
```

```python
df = pd.get_dummies(df, columns=['state'], drop_first=True)
df = pd.get_dummies(df, columns=['international_plan'], drop_first=True)
df = pd.get_dummies(df, columns=['voice_mail_plan'], drop_first=True)
```

```python
print(df.dtypes)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X = df.drop('churn', axis=1)
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
